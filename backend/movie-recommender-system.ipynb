## Convert Function
- input : genres, keywords
- Output: return a relevent list 

```python
movies.iloc[0].genres
```

```python
type(movies.iloc[0].genres)
```

```python
import ast
```

```python
def convert( obj):
    L= []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert(movies.iloc[0].genres)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head(1)
```

## top 3 cast function

```python
movies.iloc[0].cast
```

```python
def top_cast( obj):
    L= []
    count =0;
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count = count+1;
        else: 
            break
    return L
```

```python
top_cast(movies.iloc[0].cast)
```

```python
movies['cast'] = movies['cast'].apply(top_cast)
movies.head(1)
```

## Fetch director

```python
def get_director( obj):
    L= []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(get_director)
movies.head(1)
```

```python
movies.iloc[0].crew
```

## Overview Split

```python
movies['overview'].iloc[0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head(1)
```

## Trim the spaces

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['id' , 'title' , 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(1)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head(1)
```

## Stemming
- technique in NLP that reduces words to their root or base form. eg. loved , loving, loves -> love, love, love

```python
import sys
print(sys.executable)
```

```python
import nltk 
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

## Text Vectorization
- Some Techniques are :- Beg Of Words, Tf Idf, Words to veg

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

## Cosine Similarity | Cosine Distance
- Both are Inversely proportional

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

## Recommend Function

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x: x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```
